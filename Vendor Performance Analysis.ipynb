# Data Analysis

# Does purchasing in bulk reduce the unit price, and what is the optimal purchase volume for cost savings?

```python
df["UnitPurchasePrice"]=df["Total_Purchase_Dollars"]/df["Total_Purchase_Quantity"]
```

```python
df["OrderSize"]=pd.qcut(df["Total_Purchase_Quantity"],q=3,labels=["Small","Medium","Large"])
```

```python
df
```

```python
bulk_purchase_analysis=df.groupby(by="OrderSize")["UnitPurchasePrice"].mean()
```

```python
bulk_purchase_analysis
```

```python
- Vendors buying in bulk (Large Order Size) get the lowest unit price ($10.78 per unit), meaning higher margins.
- This suggests that bulk pricing strategies successfully encourage vendors to purchase in larger volumes, leading to higher overall sales 
```

# Which vendors contribute the most to total purchase dollars?

```python
vendor_performance=df.groupby(by="VendorName").agg({
    "Total_Purchase_Dollars":"sum"
}).reset_index()
vendor_performance
```

```python
vendor_performance["Purchase_Contribution%"]=(vendor_performance["Total_Purchase_Dollars"]/vendor_performance["Total_Purchase_Dollars"].sum())*100
vendor_performance
```

```python
top_vendor_performance=vendor_performance.sort_values(by="Purchase_Contribution%",ascending=False).head()
top_vendor_performance
```

```python
plt.figure(figsize=(8,8))
plt.pie(top_vendor_performance["Purchase_Contribution%"],labels=top_vendor_performance["VendorName"],autopct='%1.1f%%',
    startangle=140)
plt.title("Total Purchase Dollars by Vendor")
plt.axis('equal')  # Ensure pie is drawn as a circle
plt.tight_layout()
plt.show()
```

```python
fig,ax=plt.subplots(figsize=(10, 6))
bars=ax.bar(top_vendor_performance["VendorName"],top_vendor_performance["Purchase_Contribution%"])
ax.bar_label(bars, fmt='%.2f')
ax.set_xticklabels(top_vendor_performance["VendorName"], rotation=90)
ax.set_ylabel('Purchase_Contribution%')
ax.set_title('Vendor Contribution to Total Purchases')
plt.tight_layout()
plt.show()
```

#  Which vendors demonstrate the highest sales performance?

```python
vendor_sale=df.groupby(by="VendorName").agg({
    "Total_Sale_Dollars":"sum"
}).reset_index()
Top_vendorwise_sales=vendor_sale.sort_values(by="Total_Sale_Dollars",ascending=False).head(5)
Top_vendorwise_sales
fig,ax=plt.subplots(figure=(10,8))
bars=ax.bar(Top_vendorwise_sales["VendorName"],Top_vendorwise_sales["Total_Sale_Dollars"])
ax.bar_label(bars,fmt="%0.2f")
ax.set_xticklabels(Top_vendorwise_sales["VendorName"],rotation=90)
ax.set_ylabel('Total Sale')
ax.set_title('Top 5 vendor wise sales')
plt.tight_layout()
plt.show()
```

#  Which brands demonstrate the highest sales performance?

```python
brandwisesales=df.groupby(by="Description").agg({
    "Total_Sale_Dollars":"sum"
}).reset_index()
```

```python
topbrand=brandwisesales.sort_values(by="Total_Sale_Dollars",ascending=False).head()
topbrand
```

```python
fig,ax=plt.subplots(figure=(10,8))
bars=ax.bar(topbrand["Description"],topbrand["Total_Sale_Dollars"])
ax.bar_label(bars,fmt="%.2f")
ax.set_xticklabels(topbrand["Description"],rotation=90)
ax.set_ylabel("Total sales")
ax.set_title('Top 5 Brand wise sales')
plt.tight_layout()
plt.show()
```

# vendor with low turnover

```python
TurnoverVendor=df[df.StockTurnover<1].groupby(by="VendorName").agg({"StockTurnover":"mean"}).reset_index()
Low_Turnover=TurnoverVendor.sort_values(by="StockTurnover",ascending=True).head(5)
Low_Turnover
```

```python
 - Slow-moving inventory increases holding costs (warehouse rent, insurance, depreciation)
- Identifying vendors with low inventory turnover is critical for business efficiency, cost reduction, and profitability
```
